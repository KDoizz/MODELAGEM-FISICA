```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a função 1
def f(x):
  return(x**2 - 4) * (x**3 - 7 * x + 1)

# Definindo o intervalo

x_min = -5
x_max = 5

# Criando uma lista de pontos no intervalo
x_values = np.linspace(x_min, x_max, 1000)

# Calculando os valores da função para cada ponto
y_values = f(x_values)

# Plotando o gráfico
plt.plot(x_values, y_values)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função 1')
plt.grid()
plt.show()
```

$f'(x) = 5x^4 - 29x^2 - 8x - 28$

$f'(5) = 5\cdot5^4 - 29\cdot5^2 - 8\cdot5 - 28 = 573$

Esse valor é o coeficiente angular da reta tangente.

# O ângulo de inclinação da reta tangente é dado pela função $\arctan$ do coeficiente angular da reta tangente:
$\theta = \arctan(f'(5)) \approx 1.5508 \text{ rad} \approx 88.8711^\circ$

# Questão 2 (vale 2 pontos)
Uma empresa espacial está tentando descobrir o ponto máximo da órbita de um satélite cuja função orbital é dada por: 
$$f(x)=x^4−5x^3+10x^2−10x+5$$

Encontre o primeiro ponto de mínimo no intervalo $-10<=x<=10$.

Como sempre, você pode usar qualquer ferramenta disponível na internet para lhe auxiliar a resolver este problema. Lembre-se, contudo, que o sympy é uma boa alternativa. Em todos os casos, você precisará indicar o passo a passo da aplicação de todas as regras, técnicas, matrizes, ou tabelas que use para resolver este problema.

Em todas as questões você pode usar apenas os métodos, regras, técnicas e algoritmos que utilizamos na nossa disciplina, em sala, ou no material de apoio. Lembrem-se estamos limitados as bibliotecas sympy, numpy e matplotlib.

Para resolver esta questão você precisará de, no minimo, 4 casas decimais de precisão depois da vírgula. 

```python
import numpy as np
import plotly.graph_objects as go

def f(x):
  return x**4 - 5 * x**3 + 10 * x**2 + 5

# Preparando variáveis
x_min = -11
x_max = 11

# Gerando valores dentro do intervalo
x_values = np.linspace(x_min, x_max, 1000)
y_values = f(x_values)

# Plotando o gráfico
fig = go.Figure()
fig.add_trace(go.Scatter(x=x_values, y=f(x_values), mode='lines', name='f(x)'))
fig.update_layout(title='Gráfico das Funções f(x) ',
                  xaxis_title='x',
                  yaxis_title='f(x)',
                  legend_title='f(x)')
fig.show()

# Calculando e printando o primeiro ponto de mínimo dentro do intervalo
min_index = np.argmin(y_values)
x_min = x_values[min_index]
print(f'O primeiro ponto de mínimo está em x = {x_min:4f}')
```

# Questão 3 (vale 2 pontos)
Considerando a função $f(x,y) = \sin(xy) + \cos(x+y)$, no intervalo $0\leq x\leq 5 \space\space e \space\space 0\leq y\leq 5$ trace, usando o plotly o gráfico desta função e marque os pontos de máximo e mínimo neste intevalo.

Lembre-se o sympy é sempre uma opção e você pode usar qualquer técnica, algoritmo, regra, ou tabela, que tenha sido usada na matéria. E que precisa de apenas uma casa decimal de precisão. Uma dica importante para a solução deste problema é a função **np.unravel_index**


```python
import numpy as np
import plotly.graph_objects as go

# Definindo a função
def f(x, y):
  return np.sin(x*y) + np.cos(x + y)

# Criando os intervalos e a grade
x = np.linspace(0, 5, 100)
y = np.linspace(0, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Encontrando pontos de maximo e pontos de minimo
max_idx = np.unravel_index(np.argmax(Z), Z.shape)
min_idx = np.unravel_index(np.argmax(Z), Z.shape)
max_x, max_y, max_z = X[max_idx], Y[max_idx], Z[max_idx]
min_x, min_y, min_z = X[min_idx], Y[min_idx], Z[min_idx]

# Plotando o Gráfico
fig = go.Figure(data=[go.Surface(x=X, y=Y, z=Z)])
fig.update_layout(scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='f(X,Y)'),
                  title='Gráfico da função:')
fig.add_scatter3d(x=[max_x], y=[max_y], z=[max_z], mode='markers', marker=dict(size=10, color='red'), name='Ponto de máximo')
fig.add_scatter3d(x=[min_x], y=[min_y], z=[min_z], mode='markers', marker=dict(size=10, color='blue'), name='Ponto de mínimo')
fig.show()
```
