```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Latex, Math

def funcao_f(x):
    if x >= 2:
        raise ValueError("O valor de x não é permitido. Por favor, escolha um valor menor que 2.")
    else:
        return 1 / np.sqrt(2 - x)

def plot_funcao_f(x):
    # Plotando a função no intervalo entre ±10
    intervalo_x = np.linspace(-10, 10, 1000)
    intervalo_y = [funcao_f(i) if i < 2 else np.nan for i in intervalo_x]

    plt.plot(intervalo_x, intervalo_y, label='f(x) = 1 / √(2 - x)')
    
    # Destacando o ponto relacionado ao valor de x passado para a função
    if x < 2:
        y = funcao_f(x)
        plt.plot(x, y, 'ro', label=f'Ponto: ({x}, {y})')
    
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Gráfico da função f(x) = 1 / √(2 - x)')
    plt.grid(True)
    plt.show()

# Exemplo de uso da função e do gráfico
x = 1.5
display(Math(r'f(x) = \frac{1}{\sqrt{2-x}}'))
display(Math(r'f(%s) = %s' % (x, funcao_f(x))))
plot_funcao_f(x)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_raquete():
    # Definindo o cabo
    cabo_x = np.array([0.5, 0.5, 1.5, 1.5])
    cabo_y = np.array([-6, 0.75, 0.75, -6])

    # Definindo a área de rebatimento (circunferência)
    raio = 3.3
    theta = np.linspace(0, 2 * np.pi, 100)
    area_x = raio * np.cos(theta) + 1
    area_y = raio * np.sin(theta) + 4

    # Plotando a raquete
    plt.figure(figsize=(5, 10))
    plt.fill(area_x, area_y, 'blue', edgecolor='black') # Preenche a área de rebatimento
    plt.fill(cabo_x, cabo_y, '#deb887', edgecolor='black') # Preenche o cabo com borda preta
    plt.xlim(-5, 5)
    plt.ylim(-2, 12)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.axis('off')
    plt.show()

    plt.figure(figsize=(5, 10))
    plt.fill(area_x, area_y, 'black', edgecolor='black') # Preenche a área de rebatimento
    plt.fill(cabo_x, cabo_y, '#deb887', edgecolor='black') # Preenche o cabo com borda preta
    plt.xlim(-5, 5)
    plt.ylim(-2, 12)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.axis('off')
    plt.show()

# Chamando a função para plotar a raquete
plot_raquete()

```

```python
from IPython.display import display, Latex
def f_x(n):
    return 4 * (1 - sum(1 / (1 + 2 * i) for i in range(1,n + 1)))

n= int(input("de um valor para n: "))
fx_values = f_x(n)
latex_fx = "f(x) = 4 \\left(1 - \\sum_{i=0}^{n} \\frac{1}{1 + 2i}\\right)"

display(Latex(latex_fx))
print(f"Valores de f(x): {fx_values}")
  
```
