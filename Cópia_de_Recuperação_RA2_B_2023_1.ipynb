Derivem as funções a seguir. 
Para todas as derivadas apresente a álgebra necessária explicitando matematicamente, com latex, a aplicação de todas as regras, simplificações necessárias. Não deixe de plotar os gráficos das funçoes e das derivadas usando o matplotlib.

## 1. 
$$f(x) = x^3 + 5x^2 - 7x + 9$$

## 2.
$$g(x) = e^x \cos(x)$$


```python
# Primeira função

import numpy as np
import matplotlib.pyplot as plt

# Definindo a função
def f(x):
    return x**3 + 5*x**2 - 7*x + 9

# Definindo a derivada da função
def f_derivative(x):
    return 3*x**2 + 10*x - 7

# Criando um array de valores de x
x = np.linspace(-10, 10, 100)

# Calculando os valores correspondentes de f(x) e f'(x)
y = f(x)
y_derivative = f_derivative(x)

# Plotando o gráfico da função
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = x^3 + 5x^2 - 7x + 9')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função')
plt.grid(True)
plt.legend()

# Plotando o gráfico da derivada da função
plt.figure(figsize=(8, 6))
plt.plot(x, y_derivative, label="f'(x) = 3x^2 + 10x - 7")
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title('Gráfico da derivada da função')
plt.grid(True)
plt.legend()

# Exibindo os gráficos
plt.show()
```

# Primeira função

Passo 1: Aplicar a regra de potência para cada termo da função. A regra de potência afirma que a derivada de $x^n$ em relação a $x$ é $n \cdot x^{n-1}$.

$f'(x) = \frac{d}{dx}(x^3) + \frac{d}{dx}(5x^2) - \frac{d}{dx}(7x) + \frac{d}{dx}(9)$

Simplificando os termos, temos:

$f'(x) = 3x^2 + 10x - 7$

Portanto, a derivada da função $f(x) = x^3 + 5x^2 - 7x + 9$ em relação a $x$ é $f'(x) = 3x^2 + 10x - 7$.

```python
# Segunda função

import numpy as np
import matplotlib.pyplot as plt

# Definindo a função
def g(x):
    return np.exp(x) * np.cos(x)

# Definindo a derivada da função
def g_derivative(x):
    return np.exp(x) * (np.cos(x) - np.sin(x))

# Criando um array de valores de x
x = np.linspace(-10, 10, 100)

# Calculando os valores correspondentes de g(x) e g'(x)
y = g(x)
y_derivative = g_derivative(x)

# Plotando o gráfico da função
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='g(x) = e^x cos(x)')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.title('Gráfico da função')
plt.grid(True)
plt.legend()

# Plotando o gráfico da derivada da função
plt.figure(figsize=(8, 6))
plt.plot(x, y_derivative, label="g'(x) = e^x (cos(x) - sin(x))")
plt.xlabel('x')
plt.ylabel("g'(x)")
plt.title('Gráfico da derivada da função')
plt.grid(True)
plt.legend()

# Exibindo os gráficos
plt.show()
```

$g'(x) = \frac{d}{dx}(e^x) \cdot \cos(x) + e^x \cdot \frac{d}{dx}(\cos(x))$

A derivada de $\cos(x)$ em relação a $x$ é $-\sin(x)$.

Substituindo essas derivadas na expressão anterior, temos:

$g'(x) = e^x \cdot \cos(x) + e^x \cdot (-\sin(x))$

$g'(x) = e^x (\cos(x) - \sin(x))$

Portanto, a derivada da função $g(x) = e^x \cos(x)$ em relação a $x$ é $g'(x) = e^x (\cos(x) - \sin(x))$.

$g'(x) = \frac{d}{dx}(e^x) \cdot \cos(x) + e^x \cdot \frac{d}{dx}(\cos(x))$

Passo 2: Derivar cada uma das funções individualmente.

A derivada da função exponencial $e^x$ em relação a $x$ é igual à própria função exponencial, ou seja, $\frac{d}{dx}(e^x) = e^x$.

A derivada da função trigonométrica $\cos(x)$ em relação a $x$ é igual ao negativo da função seno, ou seja, $\frac{d}{dx}(\cos(x)) = -\sin(x)$.

Substituindo essas derivadas na expressão anterior, temos:

$g'(x) = e^x \cdot \cos(x) + e^x \cdot (-\sin(x))$

Passo 3: Avaliar a derivada no ponto $x = 5$.

Para isso, substituímos $x$ por $5$ na expressão da derivada:

$g'(5) = e^5 \cdot \cos(5) + e^5 \cdot (-\sin(5))$

Simplificando os termos, obtemos:

$g'(5) = e^5 \cdot (\cos(5) - \sin(5))$

# Derivada 2

Para calcular a derivada da função $j(x) = \sqrt{x} \ln(x)$, vamos utilizar as regras de derivação do cálculo diferencial.

Passo 1: Aplicar a regra do produto para os termos $\sqrt{x}$ e $\ln(x)$. A regra do produto afirma que a derivada do produto de duas funções é dada pela primeira função multiplicada pela derivada da segunda função, mais a segunda função multiplicada pela a derivada da primeira função.

$j'(x) = \frac{d}{dx}(\sqrt{x}) \cdot \ln(x) + \sqrt{x} \cdot \frac{d}{dx}(\ln(x))$

Passo 2: Derivar cada uma das funções individualmente.

A derivada da função raiz quadrada $\sqrt{x}$ em relação a $x$ pode ser obtida utilizando a regra da potência com expoente $\frac{1}{2}$:

$\frac{d}{dx}(\sqrt{x}) = \frac{1}{2}x^{-\frac{1}{2}}$

A derivada da função logarítmica $\ln(x)$ em relação a $x$ é dada por:

$\frac{d}{dx}(\ln(x)) = \frac{1}{x}$

Substituindo essas derivadas na expressão anterior, temos:

$j'(x) = \frac{1}{2}x^{-\frac{1}{2}} \cdot \ln(x) + \sqrt{x} \cdot \frac{1}{x}$

Simplificando e combinando os termos, obtemos:

$j'(x) = \frac{\ln(x)}{2\sqrt{x}} + \frac{1}{\sqrt{x}}$

Portanto, a derivada da função $j(x) = \sqrt{x} \ln(x)$ é $j'(x) = \frac{\ln(x)}{2\sqrt{x}} + \frac{1}{\sqrt{x}}$.

# Questão 3
Usando o python, numpy, encontre o máximo local de $$f(x)=x^2-98x+4$$ para $0 \leq x \leq 100$. 
 

```python
# Questão 3

import numpy as np

def f(x):
  return x**2 - 98*x + 4

x = np.linspace(0, 100, 1000)
y = f(x)

index = np.argmax(y)

maximum = y[index]
x_maximum = x[index]

print("Máximo local de f(x) = ", maximum)
print("Valor de x correspondente = ", x_maximum)
```
