# Questão 1 (vale até 2 pontos)
Determine a área entre as curvas $$y=xe^{-x^{2}}$$ $$y=x+1$$ $$x=2$$ e o eixo $y$, usando somente matplotlib, numpy, sympy. Trace o gráfico desta área preenchendo a área entre as curvas em uma cor diferente.



```python
# Questão 1

import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x * np.exp(-x**2)

def g(x):
    return x + 1

x = 2
y1 = f(x)
y2 = g(x)

x_interval = np.linspace(0, 2, 1000)
y1_interval = f(x_interval)
y2_interval = g(x_interval)

plt.plot(x_interval, y1_interval, label='y = x * e^(-x^2)')
plt.plot(x_interval, y2_interval, label='y = x + 1')
plt.fill_between(x_interval, y1_interval, y2_interval, where=(y1_interval > y2_interval), color='gray', alpha=0.3)
plt.ylim([0, 3])

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

# Questão 2 (vale até 2 pontos)
Resolva, algebricamente, explicitando em latex todos os passos necessários, as seguintes integrais:

##a)
$$\int \cos(3t) \sin(8t) dt $$
##b)
$$\int 4x \cos(2-3x) dx$$


# Primeira Integral:

\begin{align*}
\int \cos(3t) \sin(8t) \, dt &= \frac{1}{3} \int \cos(3t) \, d(\sin(8t)) \quad \\
&= \text{Integração por partes:} \quad u = \cos(3t), \, dv = \sin(8t) \, dt \\
&= \frac{1}{3} \left( \cos(3t) \sin(8t) - \int \sin(8t) \, d(\cos(3t)) \right) \quad \text{Aplicando a fórmula de integração por partes:} \quad du = -3 \sin(3t) \, dt, \, v = -\frac{1}{8} \cos(8t) \\
&= \frac{1}{3} \left( \cos(3t) \sin(8t) + \int \cos(8t) \, d(\cos(3t)) \right) \quad \text{Reorganizando os sinais} \\
&= \frac{1}{3} \left( \cos(3t) \sin(8t) + \int \cos(8t) \, (-3 \sin(3t)) \, dt \right) \quad \text{Aplicando a fórmula de integração por partes novamente} \\
&= \frac{1}{3} \left( \cos(3t) \sin(8t) - 3 \int \cos(8t) \sin(3t) \, dt \right) \quad \text{Reorganizando os sinais} \\
\end{align*}




\begin{align*}
\quad \text{Portanto, a solução da integral é:} \\
\int \cos(3t) \sin(8t) \, dt &= \frac{1}{3} \left( \cos(3t) \sin(8t) - 3 \int \cos(8t) \sin(3t) \, dt \right)
\end{align*}



# Segunda Integral:

\begin{align*}
\int 4x \cos(2-3x) \, dx &= \frac{1}{3} \int 4x \, d(\sin(2-3x)) \quad \text{Integração por partes:} \quad u = 4x, \, dv = \cos(2-3x) \, dx \\
&= \frac{1}{3} \left( 4x \sin(2-3x) - \int \sin(2-3x) \, d(4x) \right) \quad \text{Aplicando a fórmula de integração por partes:} \quad du = 4 \, dx, \, v = -\frac{1}{3} \sin(2-3x) \\
&= \frac{1}{3} \left( 4x \sin(2-3x) + \int \frac{4}{3} \sin(2-3x) \, dx \right) \quad \text{Reorganizando os sinais} \\
&= \frac{1}{3} \left( 4x \sin(2-3x) - \frac{4}{3} \int \sin(2-3x) \, dx \right) \quad \text{Distribuindo a constante} \\
&= \frac{1}{3} \left( 4x \sin(2-3x) - \frac{4}{3} (-\cos(2-3x)) \right) \quad \text{Aplicando a integral de } \sin(x) \text{ em relação a } x \\
&= \frac{1}{3} \left( 4x \sin(2-3x) + \frac{4}{3} \cos(2-3x) \right) \quad \text{Simplificando os termos}
\end{align*}


\begin{align*}
\text{Portanto, a solução da integral é:} \\
\int 4x \cos(2-3x) \, dx &= \frac{1}{3} \left( 4x \sin(2-3x) + \frac{4}{3} \cos(2-3x) \right)
\end{align*}


# Questão 3 (vale até 4 pontos)
Calcule o valor das integrais definidas explicitadas a seguir. Trace o gráfico das funções que foram integradas e das funções encontradas após a integração destacando, com linhas verticais, os limites de integração.

##a)
$$\int_0^2 x^2+1 \space dx$$

##b)
$$\int_{-3}^1 6x^2-5x+2 \space dx$$



```python
# Primeira Integral:

import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return x**2 + 1

a = 0
b = 2

x = np.linspace(-1, 3, 100)
y = f(x)

integral = np.trapz(y, x)

def F(x):
  return (1/3) * x**3 + x

x_integral = np.linspace(-1, 3, 100)
y_integral = F(x_integral)

plt.plot(x, y, label="Função Original")
plt.plot(x_integral, y_integral, label="Função após integração")
plt.fill_between(x_integral, y_integral)

plt.axvline(x=a, color="r")
plt.axvline(x=a, color="g")

plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico da função')
plt.legend()

plt.show()
print('Valor da integral: ', integral)
```

```python
# Segunda Integral

import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 6*x**2 - 5*x + 2

a = -3
b = 1

x = np.linspace(a, b, 100)
y = f(x)
integral = np.trapz(y, x)

plt.plot(x, y, label='Função original')
plt.fill_between(x, y, where=((x>=a) & (x<=b)), alpha=0.3)
plt.axvline(x=a, color='r', linestyle='--')
plt.axvline(x=b, color='r', linestyle='--')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico da função')
plt.show()

print("Valor da integral definida:", integral)
```
