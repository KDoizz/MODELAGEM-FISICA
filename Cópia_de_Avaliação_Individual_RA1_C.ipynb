# Questão 1: 
O tempo necessário para baixar um arquivo pode ser calculado usando logaritmos. Se a taxa de transferência de dados é de 10 Mbps e o tamanho do arquivo é de 2 GB, quanto tempo levará para baixar o arquivo? Use a fórmula:

$$
T = \frac{8 \cdot S}{R \cdot 10^6}
$$

onde $T$ é o tempo em segundos, $S$ é o tamanho do arquivo em GB e $R$ é a taxa de transferência em Mbps. Para calcular o tempo em minutos, faça $T / 60$. Não se esqueça de traçar um gráfico mostrando o o tempo de downloads para arquivos que estão entre 1 e 20 GB.

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

# Quanto tempo irá demorar para baixar o arquivo do enunciado(2GB): 2,67 segundos

tamanhos = np.arange(1, 21, 1)
taxa = 10

tempos = []
for tamanho in tamanhos:
  T = 8 * tamanho * 10**9 / (taxa * 10**6)
  tempos.append(sp.Rational(T, 60))

plt.plot(tamanhos, tempos)
plt.title('Tempo de download para diferentes tamanhos')
plt.xlabel('Tamanho do Arquivo em GB')
plt.ylabel('Tempo de download')
plt.show()
```

Usando, a seu critério, as bibliotecas do Python, crie um gráfico para cada uma das funções a seguir. No label do gráfico deverá estar a função impressa com qualidade matemática (latex) e uma label dizendo se a função é crescente ou decrescente. 

1. $e^x - 4x^2 = 0$

2. $\log_3(x^2 + 1) - \log_3(x) = 2$

```python
#@title Texto de título padrão
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

# FUNÇÃO 1 - crescente a esquerda do ponto de interseção negativo, decrescente entre os dois pontos de interseção e volta a ser crescente a direita do ponto de interseção positivo.

x = np.linspace(-2, 2, 500) # <- criando pontos entre -2 e 2
y = np.exp(x) - 4 * x**2 # <- função

plt.plot('X, Y')
plt.title('e^x − 4*x^2 = 0')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()

# FUNÇÃO 2 - crescente para x > 0 e crescente para x > 1/3

x = np.linspace(0.1, 10, 500)

y1 = np.log10(x**2 + 1)
y2 = np.log10(x) + 2

plt.plot(x, y1, label=r'$\log_{3}(x^{2} + 1)$')
plt.plot(x, y2, label=r'$\log_{3}(x) + 2$')
plt.title(r"$\log_{3}(x^{2} + 1) - \log_{3}(x) + 2$")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
