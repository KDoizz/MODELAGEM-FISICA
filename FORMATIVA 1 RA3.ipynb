Exerc√≠cio 1

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x = sp.Symbol('x')
f = 40 * x**3 + 12 * x**2 - 9 * x

integral_result = sp.integrate(f, (x, 2.5, 7))

expected_result = 24736.5

print('Resultado da integral:', integral_result.evalf())
print('Resultado esperado:', expected_result)

evaluate_f = sp.lambdify(x, f)

x_vals = np.linspace(2, 7, 100)
y_vals = evaluate_f(x_vals)

plt.plot(x_vals, y_vals)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gr√°fico da fun√ß√£o f(x)')
plt.grid(True)
plt.show()
```

Exerc√≠cio 2

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

t = sp.Symbol('t')
f = 12 * t**7 - t**2 - t + 3

integral_result = sp.integrate(f, (t, 7.5, 10))

expected_result = 30 * 10**7 - 2.5 * 10**2 - 2.5 * 10 + 7.5 - (30 * 7.5**7 - 2.5 * 7.5**2 - 2.5 * 7.5 + 7.5)

print('Resultado da integral:', integral_result)
print('Resultado esperado:', expected_result)

evaluate_f = sp.lambdify(t, f)

t_vals = np.linspace(7.5, 10, 100)
y_vals = evaluate_f(t_vals)

plt.plot(t_vals, y_vals)
plt.xlabel('t')
plt.ylabel('f(t)')
plt.title('Gr√°fico da fun√ß√£o f(t)')
plt.grid(True)
plt.show()
```

Exerc√≠cio 3

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.1, 2.5, 100)
f = (2*x**5 - x + 3) / x**2

plt.plot(x, f)
plt.fill_between(x, f, where=(x >= 1) & (x <= 2), color='gray', alpha=0.3)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gr√°fico da fun√ß√£o f(x)')
plt.grid(True)
plt.show()

a = 1
b = 2

area_algebraica = 9 - np.log(2)
num_retangulos = 100
delta_x = (b - a) / num_retangulos

x = np.linspace(a, b, num_retangulos)
f_x = (2*x**5 - x + 3) / x**2

area_riemann = np.sum(f_x * delta_x)
print('√Årea calculada usando √°lgebra:', area_algebraica)
print('√Årea calculada usando a Soma de Riemann:', area_riemann)
```

```python
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt

def integrand(x):
    return 7 * np.sin(x) - 2 * np.cos(x)

result, error = quad(integrand, 0, np.pi/2)

print('Resultado da integral:', result)

x_vals = np.linspace(0, np.pi/2, 100)
y_vals = integrand(x_vals)

plt.plot(x_vals, y_vals)
plt.fill_between(x_vals, y_vals, color='gray', alpha=0.3)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gr√°fico da fun√ß√£o')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt

def integrand(x):
    return 5 * np.sin(2 * x) + 3**x

result, error = quad(integrand, -np.pi, np.pi)

expected_result = (2 * np.sinh(np.pi * np.log(3))) / np.log(3)

print('Resultado da integral:', result)
print('Resultado esperado:', expected_result)

x_vals = np.linspace(-np.pi, np.pi, 100)
y_vals = integrand(x_vals)

plt.plot(x_vals, y_vals)
plt.fill_between(x_vals, y_vals, color='gray', alpha=0.3)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gr√°fico da fun√ß√£o')
plt.grid(True)
plt.show()
```
