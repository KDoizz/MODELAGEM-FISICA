#Questão 1 (vale até 2 pontos)
Determine a área entre as curvas definidas por $x_1=y^2-2$ e $x_2=y$. Trace o gráfico destas funções sombreando em azul claro a área entre elas. Você pode usar apenas o python, matplotlib, numpy e sympy. Neste caderno do Google Colab.

## Resposta

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, solve, integrate

# Define the variable
y = symbols('y')

# Define the functions
x1 = y**2 - 2
x2 = y

# Solve for the intersection points
intersections = solve(x1 - x2, y)

# Calculate the area between the curves
area = integrate(abs(x1 - x2), (y, intersections[0], intersections[1]))

# Generate values for y
y_vals = np.linspace(float(intersections[0]), float(intersections[1]), 1000)

# Calculate the corresponding values for x1 and x2
x1_vals = np.array([float(x1.subs(y, val)) for val in y_vals])
x2_vals = np.array([float(x2.subs(y, val)) for val in y_vals])

# Plot the functions
plt.figure(figsize=(10, 5))
plt.plot(x1_vals, y_vals, label='x1 = y^2 - 2')
plt.plot(x2_vals, y_vals, label='x2 = y')

# Correctly fill the area between the curves
for i in range(len(y_vals)):
    if x1_vals[i] > x2_vals[i]:
        plt.fill_betweenx([y_vals[i]], [x1_vals[i]], [x2_vals[i]], color='red')

# Add details to the graph
plt.title('Area between the curves x1 = y^2 - 2 and x2 = y')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

print(f'The area between the curves is {area.evalf()}')
```

Um objeto está se movendo e experimenta diferentes acelerações em diferentes tempos. Os tempos e as respectivas acelerações são:

4. No instante $t = 6$ segundos, aceleração $a = 1 m/s²$
A posição inicial do objeto é $s_0 = 0 m$ e a velocidade inicial é $v_0 = 0 m/s$.

Você precisa imprimir com qualidade matemática (latex) a função encontrada para a aceleração, a função encontrada para a posição e a posição no instante $t=4.2s$. Além, é claro, dos gráficos de aceleração e posição.

Você poderá usar apenas as bibliotecas matplotlib, numpy e sympy.

## Resposta

\begin{align*}
a(t) &=
\begin{cases}
4 \, m/s^2 & \text{para } 1 \leq t < 3 \\
-2 \, m/s^2 & \text{para } 3 \leq t < 4 \\
6 \, m/s^2 & \text{para } 4 \leq t < 6 \\
1 \, m/s^2 & \text{para } 6 \leq t
\end{cases} \\
v(t) &= \int a(t) \, dt =
\begin{cases}
4t + v_0 & \text{para } 1 \leq t < 3 \\
-2t + v(3) & \text{para } 3 \leq t < 4 \\
6t + v(4) & \text{para } 4 \leq t < 6 \\
t + v(6) & \text{para } 6 \leq t
\end{cases} \\
s(t) &= \int v(t) \, dt =
\begin{cases}
2t^2 + s_0 & \text{para } 1 \leq t < 3 \\
-t^2 + v(3)t + s(3) & \text{para } 3 \leq t < 4 \\
3t^2 + v(4)t + s(4) & \text{para } 4 \leq t < 6 \\
\frac{1}{2}t^2 + v(6)t + s(6) & \text{para } 6 \leq t
\end{cases} \\
s(4.2) &= 3(4.2)^2 + v(4)(4.2) + s(4)
\end{align*}


```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, integrate

# Define the variable
t = symbols('t')

# Define the accelerations
a1 = 4
a2 = -2
a3 = 6
a4 = 1

# Define the time intervals
t1 = np.linspace(1, 3, 100)
t2 = np.linspace(3, 4, 100)
t3 = np.linspace(4, 6, 100)
t4 = np.linspace(6, 8, 100)

# Calculate the velocity by integration
v1 = integrate(a1, t) + 0
v2 = integrate(a2, t) + v1.subs(t, 3)
v3 = integrate(a3, t) + v2.subs(t, 4)
v4 = integrate(a4, t) + v3.subs(t, 6)

# Calculate the position by integration
s1 = integrate(v1, t) + 0
s2 = integrate(v2, t) + s1.subs(t, 3)
s3 = integrate(v3, t) + s2.subs(t, 4)
s4 = integrate(v4, t) + s3.subs(t, 6)

# Plot the acceleration
plt.figure(figsize=(10, 5))
plt.plot(t1, [a1] * len(t1), label='a = 4 m/s^2')
plt.plot(t2, [a2] * len(t2), label='a = -2 m/s^2')
plt.plot(t3, [a3] * len(t3), label='a = 6 m/s^2')
plt.plot(t4, [a4] * len(t4), label='a = 1 m/s^2')
plt.title('Acceleration as a function of time')
plt.xlabel('Time (s)')
plt.ylabel('Acceleration (m/s^2)')
plt.legend()
plt.grid(True)
plt.show()

# Plot the position
plt.figure(figsize=(10, 5))
plt.plot(t1, [s1.subs(t, val) for val in t1], label='s1')
plt.plot(t2, [s2.subs(t, val) for val in t2], label='s2')
plt.plot(t3, [s3.subs(t, val) for val in t3], label='s3')
plt.plot(t4, [s4.subs(t, val) for val in t4], label='s4')
plt.title('Position as a function of time')
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, solve, integrate

# Definindo as variáveis
x = symbols('x')

# Definindo as funções
y1 = x**2
y2 = 2*x - 1

# Gerando valores para x
x_vals = np.linspace(-10, 10, 1000)

# Calculando os valores correspondentes para y1 e y2
y1_vals = [y1.subs(x, val) for val in x_vals]
y2_vals = [y2.subs(x, val) for val in x_vals]

# Plotando as funções
plt.figure(figsize=(10, 5))
plt.plot(x_vals, y1_vals, label='y = x^2')
plt.plot(x_vals, y2_vals, label='y = 2x - 1')
plt.title('Gráfico das funções y = x^2 e y = 2x - 1')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# Definindo o intervalo de integração
a = 2
b = symbols('b')

# Calculando a integral definida da diferença entre y1 e y2 no intervalo [a, b]
area = integrate(abs(y1 - y2), (x, a, b))

# Resolvendo para b tal que a área seja igual a 6
b_val = solve(area - 6, b)

print(f'O valor de b que faz a área entre as curvas ser igual a 6 é {b_val}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, solve, integrate

# Defining the variable
x = symbols('x')

# Defining the functions
y1 = x**2
y2 = 2*x - 1

# Defining the integration interval
a = 2
b = symbols('b')

# Solving for b such that the area is equal to 6
area = integrate(y1 - y2, (x, a, b))
b_val = solve(area - 6, b)

# Generating values for x
x_vals = np.linspace(a, float(b_val[0]), 1000)

# Calculating the corresponding values for y1 and y2
y1_vals = np.array([y1.subs(x, val) for val in x_vals], dtype=np.float64)
y2_vals = np.array([y2.subs(x, val) for val in x_vals], dtype=np.float64)

# Plotting the functions
plt.figure(figsize=(10, 5))
plt.plot(x_vals, y1_vals, label='y = x^2')
plt.plot(x_vals, y2_vals, label='y = 2x - 1')
plt.fill_between(x_vals, y1_vals, y2_vals, where=(y1_vals >= y2_vals), color='lightblue')
plt.title('Graph of functions y = x^2 and y = 2x - 1 with the shaded area between them')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

## Respostas

a)
\begin{align*}
\int (3x^2 + 2e^x) \, dx &= \int 3x^2 \, dx + \int 2e^x \, dx \\
&= x^3 + 2e^x + C
\end{align*}

b)
\begin{align*}
\int x \sin(x) \, dx &= uv - \int v \, du \\
&= x(-\cos(x)) - \int -\cos(x) \, dx \\
&= -x \cos(x) + \int \cos(x) \, dx \\
&= -x \cos(x) + \sin(x) + C
\end{align*}

c)
\begin{align*}
\int_{0}^{\pi} x \cos(x) \, dx &= \left[ \frac{x^2 \sin(x)}{2} + \frac{x \cos(x)}{2} \right]_{0}^{\pi} \\
&= \left( \frac{\pi^2 \sin(\pi)}{2} + \frac{\pi \cos(\pi)}{2} \right) - \left( \frac{0^2 \sin(0)}{2} + \frac{0 \cos(0)}{2} \right) \\
&= \left( \frac{\pi^2 \cdot 0}{2} + \frac{\pi \cdot (-1)}{2} \right) - \left( \frac{0^2 \cdot 0}{2} + \frac{0 \cdot 1}{2} \right) \\
&= \frac{-\pi}{2} - 0 \\
&= -\frac{\pi}{2}
\end{align*}

d)
\begin{align*}
\int x^2 e^x \, dx &=  u = x^2 \ \text{e} \ dv = e^x \, dx \\
& \Rightarrow du = 2x \, dx \ \text{e} \ v = e^x \\
& \\
\int x^2 e^x \, dx &= x^2 e^x - \int 2x e^x \, dx \\
& \\
\int 2x e^x \, dx: \\
& \\
& u = 2x \ \text{e} \ dv = e^x \, dx \\
& \Rightarrow du = 2 \, dx \ \text{e} \ v = e^x \\
& \\
\int 2x e^x \, dx &= 2x e^x - \int 2 e^x \, dx \\
&= 2x e^x - 2 \int e^x \, dx \\
&= 2x e^x - 2e^x \\
& \\
& \\
\int x^2 e^x \, dx &= x^2 e^x - (2x e^x - 2e^x) \\
&= x^2 e^x - 2x e^x + 2e^x + C
\end{align*}
