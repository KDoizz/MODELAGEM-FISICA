```python
#EXERCÍCIO 2

import math
import sympy as sy
import numpy as np
import matplotlib.pyplot as plt

#a)

h = float(input("Digite a altura do telhado:"))
print(f"O comprimento mínimo para cumprir a norma requerida é de: {h*4}")

#b)

c = float(input("Digite o comprimento do telhado:"))
print(f"A altura mínima para cumprir a norma requerida é de: {c/4}")

#c)

def plot_2_7():
  colors = ["black", "red", "purple", "blue", "yellow", "white"]
  i = 0
  fig, ax = plt.subplots()

  for comprimento in range(2, 8):
    altura = comprimento / 4

    angulo = math.atan(altura / comprimento) * 57.2958

    x = [0, comprimento, comprimento, 0]
    y = [0, 0, altura, 0]

    ax.set_title('Angulo: ' + str(round(angulo, 2)) + " graus")
    ax.plot(x, y, color=colors[i], label=f'Altura: {altura}')
    plt.legend()
    i += 1
  plt.show()

plot_2_7()

```

```python
#EXERCÍCIO 3

x = int(input("Insira um ângulo: "))
Ang = x * pi/180
def criacirculos(Ang):

  angle = np.linspace(0,2*pi,100)
  xs = cos(angle)
  ys = sin(angle)
  r1 = 0.5
  arc_angles = np.linspace(0, Ang, 20)
  arc_xs = r1 * cos(arc_angles)
  arc_ys = r1 * sin(arc_angles)

  plt.plot([cos(Ang),cos(Ang)],[0,sin(Ang)], color = 'orange')
  plt.plot([0,cos(Ang)],[0,0],color = 'blue')
  plt.plot([1, 1], [0,np.tan(Ang)], color = 'red')

  plt.plot(xs,ys, color="black")

  plt.plot([0, 0], [1, -1], color = "black")
  plt.plot([0, 1], [0, 0], color = 'black')

  plt.plot([0,cos(Ang)], [0,sin(Ang)], color = "black")
  
  plt.plot([cos(Ang),1],[sin(Ang),np.tan(Ang)],color = "black")
  plt.plot([0,cos(Ang)],[0,0],color = 'blue')

  plt.plot(arc_xs, arc_ys, color = 'green', lw = 1.5)
  plt.gca().annotate(r'$\theta$', xy=(np.cos(Ang/2) * 0.5, np.sin(Ang/2) * 0.5), xycoords='data', fontsize=12)
  plt.legend(["sin","cos","tan"])

  plt.xlim(-2,2)
  plt.ylim(-2,2)
  plt.show()

criacirculos(Ang)
```

```python
#EXERCÍCIO 4

def func_trig():
  while True:
    x = int(input("Digite uma quantidade de periodos entre 1 e 5: "))
    if x >= 1 and x <= 5:
      break
  
  sc = (2*x) - 1
  tg = x-1
  z = np.linspace(0,sc*180, 1000, endpoint=True)
  X = np.linspace(-pi,sc*pi, 1000, endpoint=True)
  C = cos(X)
  S = sin(X)
  
  plt.plot(z, C, color="blue", linewidth=2.0, linestyle="-")
  plt.title("Cosseno")
  plt.show()
  plt.plot(z, S, color="red", linewidth=2.0, linestyle="-")
  plt.title("Seno")
  plt.show()
  tgz = np.linspace(0, tg*180, 1000)
  x = np.linspace(-pi, tg*pi, 1000)
  plt.plot(tgz, (np.tan(x)),color = 'black')
  plt.title("Tangente")
  plt.show()
func_trig()

```
